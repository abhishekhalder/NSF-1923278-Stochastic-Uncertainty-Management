```python
# wind data
wind_folder = r'C:\Users\still\Documents\Github\NSF-1923278-Stochastic-Uncertainty-Management\Data\RawData\DownloadedFromERCOT\WindData'
windfiles = os.listdir(wind_folder)
wind_df = pd.DataFrame()
for file in windfiles:
    df = pd.read_csv(os.path.join(wind_folder,file),parse_dates=['INTERVAL_ENDING'])
    wind_df = pd.concat([wind_df, df])

wind_df = wind_df.sort_values(by='INTERVAL_ENDING')
```

```python
# load data
load_folder = r'C:\Users\still\Documents\Github\NSF-1923278-Stochastic-Uncertainty-Management\Data\RawData\DownloadedFromERCOT\LoadData'
loadfiles = os.listdir(load_folder)
load_df = pd.DataFrame()
for file in loadfiles:
#     print(file)
    df = pd.read_csv(os.path.join(load_folder,file),parse_dates=['IntervalEnding'])
    load_df = pd.concat([load_df, df])

load_df.sort_values(by='IntervalEnding')
# load_df.to_csv('load_time_series.csv', index=False)
```

```python
load_E_df = load_df[load_df['Model']=='E']
load_E_df = load_E_df.sort_values(by='IntervalEnding')
load_A6_df = load_df[load_df['Model']=='A6']
load_A6_df = load_A6_df.sort_values(by='IntervalEnding')
```

```python
load_E_df.to_csv('load_modelE_time_series.csv', index=False)
load_A6_df.to_csv('load_modelA6_time_series.csv', index=False)
```
